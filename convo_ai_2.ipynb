```python
from sklearn.datasets import load_iris
import pandas as pd

data=load_iris()
df_cpu=pd.DataFrame(data.data,columns=data.feature_names)
labels=data.target

import cudf
df_gpu=cudf.DataFrame.from_pandas(df_cpu)
df_gpu["label"]=labels
print(df_gpu.head())
```

```python
import time
start=time.time()
print(df_gpu.mean())
print("Time",time.time()-start)
```

```python
import time
start=time.time()
print(df_gpu.std())
print(df_gpu.var())
print("Time",time.time()-start)
```

```python
import time
start=time.time()
print(df_gpu.median())
print(df_gpu.quantile((0.25,0.75)))
print("Time",time.time()-start)
```

```python
import time
start=time.time()
print(df_gpu.drop("label",axis=1).corr())
print("Time",time.time()-start)
```
